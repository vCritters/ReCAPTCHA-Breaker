```python
import os
from keras.preprocessing.image import ImageDataGenerator
```

```python
dataset_path = '/Users/SmashBros/Documents/GitHub/ReCAPTCHA-Breaker/datasets'
CarData_folder = dataset_path + '/CarData'

# train folder
train_folder = CarData_folder + '/TrainImages/'

# split training data into negative and postive folders
# negative = not a car
# positive = is a car
negative_folder = train_folder + '/negative/'
positive_folder = train_folder + '/postive/'

# test folder
test_folder = CarData_folder + '/TestImages/unknown'
```

```python
os.listdir(test_folder)
```

```python
train_datagen = ImageDataGenerator(rescale            = 1/255,
                                   shear_range        = 0.2,
                                   zoom_range         = 0.2,
                                   horizontal_flip    = True,
                                   rotation_range     = 40,
                                   width_shift_range  = 0.2,
                                   height_shift_range = 0.2)

test_datagen = ImageDataGenerator(rescale = 1/255)

training_set = train_datagen.flow_from_directory(train_folder,
                                   color_mode = 'grayscale',
                                   target_size = (64, 64),
                                   batch_size  = 32,
                                   class_mode  = 'binary')


test_set = test_datagen.flow_from_directory(test_folder,
                                   color_mode = 'grayscale',
                                   target_size = (64, 64),
                                   batch_size = 32,
                                   class_mode = 'binary')
```
